#### Exercise 2.1: Compute Intrinsics
NOTE: this exercise needs to be completed after the exercises described in `camera_intrinsics.ipynb`

In the file, `camera_calibration.py`, implement the function `compute_intrinsics` to compute the camera intrinsic matrix $K$ given a list of homographies. Then, run the code below to display the result.

```python
# Compute world coordinate points
X_w, Y_w, Z_w = generate_chessboard_3D_world_coordinates()

# For each image, compute the pixel coordinates of the corners
u_meas = []; v_meas = []
for img_path in IMG_PATHS:
    u, v = generate_chessboard_2D_pixel_coordinates(img_path)
    u_meas.append(u); v_meas.append(v)

# Compute homography for each image.
homographies = []
for u, v in zip(u_meas, v_meas):
    homographies.append(compute_homography(u, v, X_w, Y_w))

# Compute intrinsics from homographies.
K = compute_intrinsics(homographies)
print(K)
```
